```python
%matplotlib inline
import matplotlib.pyplot as plt
import matplotlib
```

```python
import nibabel as nib
import skimage
from skimage import filters, feature, morphology, exposure
import scipy
```

```python
img = nib.load('s28_brain.nii.gz')
img = img.get_data()
print img.shape
```

```python
slice1 = img[120, :, :]
slice2 = img[ 80, :, :]
```

```python
fig, (ax1, ax2) = plt.subplots(1,2)
ax1.imshow(slice1.T, cmap='gray', origin="lower")
ax2.imshow(slice2.T, cmap='gray', origin="lower")
```

```python
slice_canny = feature.canny(slice1, sigma=3)
```

```python
fig, (ax1, ax2) = plt.subplots(1,2)
ax1.imshow(slice1.T, cmap='gray', origin="lower")
ax2.imshow(slice_canny.T, cmap='gray', origin="lower")
```
